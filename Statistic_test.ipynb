# ATAL statistic test

```python
def compute_cell(row,metric):
    
    row = ast.literal_eval(row)
    tp_h, tn_h, tp_m, tn_m, fn_m, fp_m = row['TP_H'], row['TN_H'], row['TP_M'], row['TN_M'], row['FN_M'], row['FP_M']
    N = tp_h+ tn_h+ tp_m+tn_m+ fn_m+ fp_m
    
    if metric == 'acc+':
        eva_result = (tp_h+tp_m+tn_h+tn_m)/N
        
    if metric == 'yield':
        eva_result = (tp_h+tp_m)/(tp_h+tp_m+fn_m)
        
    if metric == 'burden':
        eva_result = (tp_h+tn_h+tp_m+fp_m)/N
        
    if metric == 'precision':
        eva_result = (tp_h+tp_m)/(tp_h+tp_m+fp_m)
        
    if metric == 'macro_recall':
        recall_1 = (tp_h+tp_m)/(tp_h+tp_m+fn_m)
        recall_2 = (tn_h+tn_m)/(tn_h+tn_m+fp_m)
        
        eva_result = (recall_1+recall_2)/2
        
    if metric =='U19':
        yield_ = (tp_h+tp_m)/(tp_h+tp_m+fn_m)
        burden = (tp_h+tn_h+tp_m+fp_m)/N
        
        eva_result = (19*yield_+(1-burden))/20
        
    if metric == 'coverage':
        eva_result = (tp_h)/(tp_h+tp_m+fn_m)
        
    if metric == 'macro_f_score':
        
        recall_1 = (tp_h+tp_m)/(tp_h+tp_m+fn_m)
        recall_2 = (tn_h+tn_m)/(tn_h+tn_m+fp_m)
        
        precision_1 = (tp_h+tp_m)/(tp_h+tp_m+fp_m)
        precision_2 = (tn_h+tn_m)/(tn_h+tn_m+fn_m)
        
        fscore_1 = 2*precision_1*recall_1/(precision_1+recall_1)
        fscore_2 = 2*precision_2*recall_2/(precision_2+recall_2)
        
        eva_result = (fscore_1+fscore_2)/2
        
    
    
    return eva_result

def compute_all(df,metric):
    
    row_num = df.shape[0]
    col_num = df.shape[1]
    
    avg_accuracies = np.zeros((row_num,col_num))
    
#     print(row_num,col_num)
    
    
    for i in range(col_num):
        col = df.iloc[:,i].apply(compute_cell,metric=metric).tolist()
        avg_accuracies[:,i] = np.array(col)
        
    return avg_accuracies
```

```python
area = 990
metric = 'acc+'
x_axis = [10+i*10 for i in range(100)]
  
aulcs = {}
    
representation = 'roberta-base'
datasets = ['Longer_MovieReview','Longer_MultiDomainCustomerReview','Longer_BlogAuthorGender','Guardian2013',
           'AdditionalCustomerReview','MovieReviewSubjectivity','ag_news','dbpedia']

for dataset in datasets:

    df = pd.read_csv('./raw_%s_%s_neg.csv_%s_pos.csv_uncertainty_gridsearch.csv'%(representation,dataset,dataset),index_col=[0])
    avg_score = np.mean(compute_all(df,metric),axis=1)
    area_ = trapz(avg_score,x_axis, dx=1)/area
    if dataset not in aulcs.keys():
        aulcs[dataset] = []
        aulcs[dataset].append(area_)


    df_tuned = pd.read_csv('./raw_%s_%s_tuned_neg.csv_%s_tuned_pos.csv_uncertainty_gridsearch.csv'%(representation,dataset,dataset),index_col=[0])
    avg_score_tuned = np.mean(compute_all(df_tuned,metric),axis=1)
    area_tuned = trapz(avg_score_tuned,x_axis, dx=1)/area
    aulcs[dataset].append(area_tuned)

    df_adaptive = pd.read_csv('./raw_roberta-base_%s_uncertainty_gridsearch.csv'%(dataset),index_col=[0])
    avg_score_adaptive = np.mean(compute_all(df_adaptive,metric),axis=1)
    area_adaptive = trapz(avg_score_adaptive,x_axis, dx=1)/area
    aulcs[dataset].append(area_adaptive)
```
